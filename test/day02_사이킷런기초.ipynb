### 첫번째 머신러닝 만들어 보기 - 붓꽃 품종 예측하기
- DecisionTreeClassifier 분류용
- DecisionTreeRegressor 회귀용
- train_test_split 학습데이터와 테스트데이터 나눠주는 라이브러리

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
load_iris()
```

```python
import pandas as pd
```

```python
iris = load_iris()
iris
```

```python
iris['data']
```

```python
iris.data
```

```python
iris_data = iris.data
iris_label = iris.target
```

```python
target_names = iris.target_names
target_names[iris_label[0]]
```

```python
iris_df = pd.DataFrame(data=iris_data,columns=iris.feature_names)
```

```python
iris_df['label'] = iris_label
iris_df.sample(3)
```

```python
# test_size : test와 train 비율! 둘중 하나만 써주면 된다
# random_state : seed 값 넣어주는거랑 똑같다 
X_train,X_test,y_train,y_test = train_test_split(iris_data,iris_label,test_size=0.2,random_state=11)
```

```python
# 모델
dt_clf = DecisionTreeClassifier(random_state=11)
# 데이터 넣어서 학습
dt_clf.fit(X_train,y_train)
# 잘만들어 졌는지 확인 (예측)
pred = dt_clf.predict(X_test)
```

```python
# metrics : 측정하는 모든 것과 관련되어 있다
from sklearn.metrics import accuracy_score
accuracy_score(y_test,pred)
```
